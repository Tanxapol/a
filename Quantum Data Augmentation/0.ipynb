```python
import matplotlib.pyplot as plt

# unused but required import for doing 3d projections with matplotlib < 3.2
import mpl_toolkits.mplot3d  # noqa: F401

from sklearn.decomposition import PCA
from matplotlib.lines import Line2D

fig = plt.figure(1, figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d", elev=-150, azim=110)

X_reduced = PCA(n_components=3).fit_transform(X)
X_reduced2 = PCA(n_components=3).fit_transform(df_train3[['sepal length', 'sepal width', 'petal length', 'petal width']])
y2 = df_train3['label'].values
scatter = ax.scatter(
    X_reduced[:, 0],
    X_reduced[:, 1],
    X_reduced[:, 2],
    c=y,
    s=40,
)
scatter2 = ax.scatter(
    X_reduced2[:, 0],
    X_reduced2[:, 1],
    X_reduced2[:, 2],
    c=y2,
    s=40,
    label='Generated Data',
)

ax.set(
    title="First three PCA dimensions",
    xlabel="1st Eigenvector",
    ylabel="2nd Eigenvector",
    zlabel="3rd Eigenvector",
)
ax.xaxis.set_ticklabels([])
ax.yaxis.set_ticklabels([])
ax.zaxis.set_ticklabels([])

# Add a legend
handles, labels = scatter.legend_elements()
legend1 = ax.legend(
    handles,
    labels,
    loc="upper right",
    title="Classes",
)
ax.add_artist(legend1)

# Example: custom colors for real and generated data

# Remove previous legends to avoid duplicates
ax.get_legend().remove()

# Custom color mapping for classes (real data)
class_colors = {0: 'tab:blue', 1: 'tab:orange', 2: 'tab:green'}
colors_real = [class_colors[label] for label in y]
scatter.remove()
scatter = ax.scatter(
    X_reduced[:, 0],
    X_reduced[:, 1],
    X_reduced[:, 2],
    c=colors_real,
    s=40,
    label='Real Data',
)

# Custom color mapping for generated data (use different shades or symbols)
gen_class_colors = {7: 'tab:purple', 8: 'tab:red', 9: 'tab:brown'}
colors_gen = [gen_class_colors[label] for label in y2]
scatter2.remove()
scatter2 = ax.scatter(
    X_reduced2[:, 0],
    X_reduced2[:, 1],
    X_reduced2[:, 2],
    c=colors_gen,
    s=40,
    marker='^',
    label='Generated Data',
)

# Custom legend
legend_elements = [
    Line2D([0], [0], marker='o', color='w', label='Setosa (Real)', markerfacecolor='tab:blue', markersize=8),
    Line2D([0], [0], marker='o', color='w', label='Versicolor (Real)', markerfacecolor='tab:orange', markersize=8),
    Line2D([0], [0], marker='o', color='w', label='Virginica (Real)', markerfacecolor='tab:green', markersize=8),
    Line2D([0], [0], marker='^', color='w', label='Setosa (Gen)', markerfacecolor='tab:purple', markersize=8),
    Line2D([0], [0], marker='^', color='w', label='Versicolor (Gen)', markerfacecolor='tab:red', markersize=8),
    Line2D([0], [0], marker='^', color='w', label='Virginica (Gen)', markerfacecolor='tab:brown', markersize=8),
]
ax.legend(handles=legend_elements, loc='upper left', title='Class / Source')

plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# Split the data into training and testing sets
X_train = df_train3[['sepal length', 'sepal width', 'petal length', 'petal width']]
y_train = df_train3['label'].replace({7: 0, 8: 1, 9: 2})

X_test = df[['sepal length', 'sepal width', 'petal length', 'petal width']]
y_test = df['label']
```

```python
clf = RandomForestClassifier(random_state=42)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)


print("Random Forest Classifier accuracy:", accuracy_score(y_test, y_pred))

print("Classification Report (Test on real Iris):\n")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
clf = SVC(kernel='linear',C=1.0 , random_state=42)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)


print(" SVC Classifier accuracy:", accuracy_score(y_test, y_pred))

print("Classification Report (Test on real Iris):\n")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler

# === โหลดข้อมูล Iris จริง ===
X_real, y_real = load_iris(return_X_y=True)
df_real = pd.DataFrame(X_real, columns=["sepal length", "sepal width", "petal length", "petal width"])
df_real["label"] = y_real
df_real["source"] = "Real"

# === โหลดข้อมูลที่ generate ด้วย VQC ===
# (จากขั้นตอนก่อนหน้านี้)
X_synth = df_train3[['sepal length', 'sepal width', 'petal length', 'petal width']].values
y_synth = df_train3['label'].replace({7: 0, 8: 1, 9: 2}).values

df_synth = pd.DataFrame(X_synth, columns=["sepal length", "sepal width", "petal length", "petal width"])
df_synth["label"] = y_synth
df_synth["source"] = "Synthetic"

# === รวมข้อมูล ===
df_all = pd.concat([df_real, df_synth], ignore_index=True)

# === Standardize ก่อนทำ PCA / t-SNE ===
features = ["sepal length", "sepal width", "petal length", "petal width"]
X_all_std = StandardScaler().fit_transform(df_all[features])

# === PCA 2D ===
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_all_std)
df_all["PCA1"] = X_pca[:, 0]
df_all["PCA2"] = X_pca[:, 1]

# === t-SNE 2D (ช้ากว่า แต่เห็น cluster ชัดขึ้น) ===
tsne = TSNE(n_components=2, perplexity=30, n_iter=1000, random_state=42)
X_tsne = tsne.fit_transform(X_all_std)
df_all["TSNE1"] = X_tsne[:, 0]
df_all["TSNE2"] = X_tsne[:, 1]

# === Plot: PCA ===
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
for source in ["Real", "Synthetic"]:
    for label in [0, 1, 2]:
        subset = df_all[(df_all["source"] == source) & (df_all["label"] == label)]
        plt.scatter(subset["PCA1"], subset["PCA2"], label=f"{source} - Class {label}", alpha=0.6)
plt.title("PCA: Real vs Synthetic Data")
plt.xlabel("PCA1")
plt.ylabel("PCA2")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# === Plot: t-SNE ===
plt.subplot(1, 2, 2)
for source in ["Real", "Synthetic"]:
    for label in [0, 1, 2]:
        subset = df_all[(df_all["source"] == source) & (df_all["label"] == label)]
        plt.scatter(subset["TSNE1"], subset["TSNE2"], label=f"{source} - Class {label}", alpha=0.6)
plt.title("t-SNE: Real vs Synthetic Data")
plt.xlabel("TSNE1")
plt.ylabel("TSNE2")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
